```python
X = df_train.iloc[:, 1:]
Y  = df_train.iloc[:, 0]

print(X.shape, Y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
x_scaled = pd.DataFrame(x_scaled, columns=X.columns)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test  = train_test_split(x_scaled, Y, test_size=0.3)
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
from sklearn import svm
from sklearn.svm import SVC

svm = SVC()
svm.fit(x_train, y_train)
svm_train_pred = svm.predict(x_train)
svm_test_pred = svm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("D SVM train 정확도 : {0:.3f}".format(accuracy_score(y_train, svm_train_pred)))
print("D SVM test  정확도 : {0:.3f}".format(accuracy_score(y_test, svm_test_pred)))
```

```python
from sklearn.metrics import confusion_matrix

print("D SVM train\n", confusion_matrix(y_train, svm_train_pred))
print()
print("D SVM test\n", confusion_matrix(y_test, svm_test_pred))
```
